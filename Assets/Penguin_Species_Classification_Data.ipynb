```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Cargar el archivo CSV
df = pd.read_csv('/content/penguins_size.csv')

# Mostrar las primeras filas del DataFrame
print(df.head())

# Información general del DataFrame
print(df.info())

# Estadísticas descriptivas
print(df.describe())

# Verificar valores nulos
print(df.isnull().sum())

# Limpieza de datos: Eliminar filas con valores nulos
df_cleaned = df.dropna()

# Análisis inicial
print(df.head())
print(df.info())
print(df.describe())

# Limpieza de datos
df = df.dropna()  # Eliminar filas con valores faltantes
```

# ***Análisis exploratorio: Preprocesamiento de datos.***

---

# Gráfico de barras: Distribución de especies de pinguinos.

```python
# Análisis de especies
species_counts = df_cleaned['species'].value_counts()
print(species_counts)

# Gráfico de barras de especies
plt.figure(figsize=(8, 6))
sns.countplot(x='species', data=df_cleaned)
plt.title('Distribución de Especies de Pingüinos')
plt.xlabel('Especies')
plt.ylabel('Cantidad')
plt.show()
```

# Gráfico de barras: Distribución sexos de los pinguinos.

```python
# Análisis de sexo
sex_counts = df_cleaned['sex'].value_counts()
print(sex_counts)

# Gráfico de barras de sexo
plt.figure(figsize=(8, 6))
sns.countplot(x='sex', data=df_cleaned)
plt.title('Distribución de Sexo en Pingüinos')
plt.xlabel('Sexo')
plt.ylabel('Cantidad')
plt.show()
```

# Gráfico de barras: Distribución de Pinguinos por isla.

```python
# Análisis de islas
island_counts = df_cleaned['island'].value_counts()
print(island_counts)

# Gráfico de barras de islas
plt.figure(figsize=(8, 6))
sns.countplot(x='island', data=df_cleaned)
plt.title('Distribución de Pingüinos por Isla')
plt.xlabel('Isla')
plt.ylabel('Cantidad')
plt.show()
```

# Histograma: Distribución de longitus del pico.

```python
# Análisis de la longitud del pico
culmen_stats = df_cleaned['culmen_length_mm'].describe()
print(culmen_stats)

# Histograma de longitud del pico
plt.figure(figsize=(8, 6))
sns.histplot(df_cleaned['culmen_length_mm'], kde=True)
plt.title('Distribución de la Longitud del Pico')
plt.xlabel('Longitud del Pico (mm)')
plt.ylabel('Frecuencia')
plt.show()
```

# ***Análisis exploratorio: Procesamiento de datos.***

---

#Histograma: Distribución de la masa corporal por especie.

```python
# Análisis de la masa corporal por especie
body_mass_by_species = df.groupby('species')['body_mass_g'].describe()
print("\nEstadísticas descriptivas de la masa corporal por especie:")
print(body_mass_by_species)

# 1. Visualización de la masa corporal por especie
g = sns.FacetGrid(df, col='species', height=5, aspect=1)
g.map(sns.histplot, 'body_mass_g', kde=True)
g.set_axis_labels('Masa Corporal (g)', 'Frecuencia')
plt.suptitle('Distribución de la Masa Corporal por Especie (Facet Grid)')
plt.show()
```

#Gráfico Dispersión: Correlación entre longitud del pico y la masa corporal.

```python
# Análisis de la longitud del pico
culmen_length_stats = df['culmen_length_mm'].describe()
print("Estadísticas descriptivas de la longitud del pico:")
print(culmen_length_stats)

# Análisis de la masa corporal
body_mass_stats = df['body_mass_g'].describe()
print("\nEstadísticas descriptivas de la masa corporal:")
print(body_mass_stats)

# 2. Correlación entre longitud del pico y masa corporal
plt.figure(figsize=(10, 6))
sns.scatterplot(x='culmen_length_mm', y='body_mass_g', hue='species', data=df)
plt.title('Correlación entre Longitud del Pico y Masa Corporal')
plt.xlabel('Longitud del Pico (mm)')
plt.ylabel('Masa Corporal (g)')
plt.show()
```

#Gráfico Barras: Distribución de especies por isla.

```python
# Análisis de la distribución de especies por isla
species_by_island = df.groupby(['island', 'species']).size().unstack()
print("Distribución de especies por isla:")
print(species_by_island)

# 3. Distribución de especies por isla
plt.figure(figsize=(10, 6))
sns.countplot(x='island', hue='species', data=df)
plt.title('Distribución de Especies por Isla')
plt.xlabel('Isla')
plt.ylabel('Cantidad')
plt.show()
```

#Gráfico Caja: Diferencias en masa corporal entre machos y hembras.

```python
# Análisis de la masa corporal por especie y sexo
body_mass_by_species_sex = df.groupby(['species', 'sex'])['body_mass_g'].describe()
print("Estadísticas descriptivas de la masa corporal por especie y sexo:")
print(body_mass_by_species_sex)

# 4. Diferencias en masa corporal entre machos y hembras
plt.figure(figsize=(10, 6))
sns.boxplot(x='species', y='body_mass_g', hue='sex', data=df)
plt.title('Diferencias en Masa Corporal entre Machos y Hembras')
plt.xlabel('Especie')
plt.ylabel('Masa Corporal (g)')
plt.show()
```

# Gráfico Pastel: Distribución de especies.

```python
# Análisis de la distribución de especies
species_counts = df['species'].value_counts()
print("Frecuencia de especies:")
print(species_counts)

# 6. Diagrama de pastel de la distribución de especies
species_counts = df['species'].value_counts()
plt.figure(figsize=(8, 8))
plt.pie(species_counts, labels=species_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Distribución de Especies')
plt.show()
```

#Matriz De Correlación: Coeficientes de correlación entre varias variables.

```python
# Seleccionar solo las columnas numéricas para la matriz de correlación
numeric_columns = df_cleaned.select_dtypes(include=['float64', 'int64'])

# Calcular la matriz de correlación
correlation_matrix = numeric_columns.corr()

# Visualizar la matriz de correlación
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Matriz de Correlación')
plt.show()
```
